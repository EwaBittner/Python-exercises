### Zad. 2. Napisz funkcję, która podniesie do wielkiej litery nazwy kolumn. Zmień je w datasecie mat. (dla chętnych: zrobienie tego za pomocą lambdy)

```python
def to_title(x):
    return x.title()
```

```python
mat.columns = [to_title(col) for col in mat]
```

### Zad. 3. Spróbuj zaaplikować te funkcję również na na kolumnach Fjob oraz Mjob dla zestawu mat.

```python
mat["Mjob"] = mat["Mjob"].apply(to_title)
por["Fjob"] = por["Fjob"].apply(to_title)
```

### Zad. 4. Napisz funkcje, ktora zwroci True albo False jeśli dany student ma powyżej 16 lat. Nazwij ja jak chcesz.

```python
def calculate_age(x):
    if x > 16:
        return True
    else:
        return False
```

### Zad. 5. Uzywajac powyzszej funkcji dodaj do wybranego zestawu danych kolumne z informacja, ktora jest wynikiem funkcji.

```python
mat["Over_16"] = mat.Age.apply(calculate_age)
```
