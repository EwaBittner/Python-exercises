**Zad.8. Jak nazywa sie osoba z najwieksza wyplata (z benefitami)?**

```python
data['EmployeeName'][data['TotalPayBenefits'] == data['TotalPayBenefits'].max()]
```

**Zad.9. Jak nazywa sie osoba z najnizsza wyplata (z benefitami)? Ile zarabia?**

```python
data[['EmployeeName', 'TotalPayBenefits']][data['TotalPayBenefits'] == data['TotalPayBenefits'].min()]
```

**Zad.10. Jaka jest srednia BasePay dla wszystkich pracownikow dla kazdego roku?(2011-2014)**

```python
import numpy as np 

data['BasePay'] = data['BasePay'].apply(lambda x: np.NaN if x == "Not Provided" else x).astype("float")
data.groupby(by='Year')['BasePay'].mean()
```
