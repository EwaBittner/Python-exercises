```python
import pandas as pd
import matplotlib.pyplot as plt  

#df = pd.read_csv("C:\\Users\\Milaan\\01_Learn_Python4Data\\11_Python_Matplotlib_Module\\company_sales_data.csv")
df = pd.read_csv('company_sales_data.csv')
profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()
plt.plot(monthList, profitList, label = 'Month-wise Profit data of last year')
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(monthList)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

Zad.3. Zmodyfikuj wykres w taki sposb aby wyglda nastepujco:
    


```python
plt.plot(monthList, profitList,color="red",linestyle="--",linewidth=4,marker="o",markersize=7,markerfacecolor='k')
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(monthList)
plt.legend(['Profit data of last year'],loc="lower right")
plt.title('Company sales data of last year')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

Zad.4. Wywietlaj liczb sprzedanych jednostek miesicznie dla kadego produktu za pomoc wykresw wielowierszowych. (tj. oddzielna fabua dla kadego produktu ).

Wykres powinien wyglda tak:
    


```python
plt.plot(monthList,df.facecream,color="blue",linestyle="-",linewidth=4,marker="o",markersize=8,markerfacecolor='blue')
plt.plot(monthList,df.facewash,color="brown",linestyle="-",linewidth=4,marker="o",markersize=8,markerfacecolor='brown')
plt.plot(monthList,df.toothpaste,color="green",linestyle="-",linewidth=4,marker="o",markersize=8,markerfacecolor='green')
plt.plot(monthList,df.bathingsoap,color="red",linestyle="-",linewidth=4,marker="o",markersize=8,markerfacecolor='red')
plt.plot(monthList,df.shampoo,color="purple",linestyle="-",linewidth=4,marker="o",markersize=8,markerfacecolor='purple')
plt.plot(monthList,df.moisturizer,color="orange",linestyle="-",linewidth=4,marker="o",markersize=8,markerfacecolor='orange')
plt.xlabel("Month Number")
plt.ylabel("Sales Unit in Number")
plt.title("Sales Data")
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.legend(["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"])
plt.show()
```

Zad.5. Za pomoc subplotw sprbuj wykona nastpujce wykresy:
    


```python
fig, ax = plt.subplots(2,sharex=True)
ax[0].plot(monthList,df.bathingsoap,color="k",linestyle="-",linewidth=4,marker="o",markersize=8,markerfacecolor='k')
ax[1].plot(monthList,df.facewash,color="red",linestyle="-",linewidth=4,marker="o",markersize=8,markerfacecolor='red')
ax[0].set_title("Sales data of Bathingsoap")
ax[1].set_title("Sales data of a facewash")
ax[1].set(ylabel="Sales units in number",xlabel="Month number")
plt.show()
```
